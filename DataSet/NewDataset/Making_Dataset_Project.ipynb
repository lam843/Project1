```python
import pandas as pd

sheet_name = ['/content/drive/MyDrive/saison_file.csv', '/content/drive/MyDrive/precoce_file.csv']

df_saison = pd.read_csv(sheet_name[0])
df_precoce = pd.read_csv(sheet_name[1])

# Add 'variety' column with value 1 to df_saison
df_saison['variety'] = 1

# Add 'variety' column with value 2 to df_precoce
df_precoce['variety'] = 2

# Merge the two datasets
df = pd.concat([df_saison, df_precoce], ignore_index=True)
```

```python
df
```

```python
'''poids=['PFr1', 'PFr2', 'PFr3', 'PFr4', 'PFr5', 'PFr6']
total_weight = df[poids].sum(axis=1)
df['Total_Poids'] = total_weight'''
import pandas as pd

poids = ['PFr1', 'PFr2', 'PFr3', 'PFr4', 'PFr5', 'PFr6']

# Convert the columns to numeric
df[poids] = df[poids].apply(pd.to_numeric, errors='coerce')

df['Total_Poids'] = df[poids].sum(axis=1, skipna=True)
```

```python
df.drop([ 'DFr1', 'DFr2', 'DFr3', 'DFr4','DFr6','DFr5'], axis=1, inplace=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Save the training and testing data 
train_data.to_csv('/content/drive/MyDrive/train_dataset.csv', index=False)
test_data.to_csv('/content/drive/MyDrive/test_dataset.csv', index=False)
```
