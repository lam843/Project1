# Stats for branches

```python
def stats_about_branches(df):
    df_ab = df[['DChar3', 'DChar4', 'DChar5']]
    missing_values = df_ab.isnull().sum()
    
    plt.figure(figsize=((20,10)))

    plt.subplot(1,2,1)
    total_missing = df_ab.isnull().sum().sum()
    plt.pie(missing_values, labels=['DChar3', 'DChar4', 'DChar5'], autopct='%1.1f%%')
    plt.title(f'Total Missing Values: {total_missing}')

    plt.subplot(1,2,2)
    missing_values = df_ab.isnull().sum()
    missing_values.plot(kind='bar')
    plt.xlabel('Columns')
    plt.ylabel('Number of missing values')
    plt.title('Bar Chart of Missing Values by Column')


    plt.show()
    
```

```python
stats_about_branches(df_saison)
stats_about_branches(df_precoce)
```

```python
def show_heat_map(df):
    sns.heatmap(df[['DChar3', 'DChar4', 'DChar5']].isnull(), cmap='Blues', cbar=False) # missing values in blue
    plt.xlabel('Columns')
    plt.title('Heatmap of Missing Values by Column')
    plt.show()
```

```python
show_heat_map(df_saison)
show_heat_map(df_precoce)
```

# Stats about Fruits


```python
def stats_about_fruites(df):
    df_ab_fr = df[['PFr3', 'PFr4','PFr6', 'PFr5']]
    missing_values = df_ab_fr.isnull().sum()
    plt.figure(figsize=((25,5)))

    plt.subplot(1,2,1)
    total_missing = df_ab_fr.isnull().sum().sum()
    plt.pie(missing_values, labels=['PFr3', 'PFr4','PFr6', 'PFr5'], autopct='%1.1f%%')
    plt.title(f'Total Missing Values: {total_missing}')
    plt.show()

    plt.subplot(1,2,2)
    missing_values.plot(kind='bar')

    plt.xlabel('Columns')
    plt.ylabel('Number of missing values')
    plt.title('Bar Chart of Missing Values by Column')

    plt.show()
    
```

```python
stats_about_fruites(df_saison)
stats_about_fruites(df_precoce)
```

```python
def show_heat_map_fruites(df):
    cols = df[['PFr3', 'PFr4','PFr6', 'PFr5']].isnull()
    sns.heatmap(cols, cmap='Blues', cbar=False) # missing values in blue

    plt.xlabel('Columns')
    plt.title('Heatmap of Missing Values by Column')

    # Display the plot
    plt.show()
```

```python
show_heat_map_fruites(df_precoce)
show_heat_map_fruites(df_saison)
```
